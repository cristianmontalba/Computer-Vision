# 2) Rectas y líneas




*   Encuentre la ecuación de la recta que pasa por los puntos (13 , 15.2) y (-7.2 , 18.2)
*   Encuentre la intersección de la recta $y=12x+13$ con la recta $y=-x/7$



```python
# Calcular la pendiente m

x1 = 13
y1 = 15.2
x2 = -7.2
y2 = 18.2
m = (y2 - y1) / (x2 - x1)
print("Pendiente (m):", m)

# Intersección
n = y2 - m * x2
print("Intersección (n):", n)

# Ecuación de la recta
print("Ecuación de la recta: y = {:.6f}x + {:.6f}".format(m, n))

# Intersección: y = 12x + 13 con la recta y = -x/7

# 12x + 13 = -x/7
# (12x + x/7 + 13 = 0) * 7
# 7*12x + 7*(x/7) + 7*13 = 0
# 84x + x + 91 = 0
# 85x + 91 = 0

x = -91 / 85
y = -x / 7  # La segunda ecuación para encontrar y

print("x =", x)
print("y =", y)
print("Intersección: ({:.6f}, {:.6f})".format(x, y))
```

# 8) Encuentre mc, el centro del cuadrilátero y despliéguelo

(2 Puntos)

*Usando la metodología vista en clases -con coordenadas homogénas-, encuentre el centro del cuadrilatero como la intersección de la recta que pasa por (m1 y m4) con la recta que pasa por (m2 y m3).*

*Debe mostrar los cuatro puntos del cuadrilátero, la línea del cuadrilátero, y el centro del cuadrilátero sobre la imagen de John Lennon.*


```python
# PREGUNTA 8: ESCRIBA CODIGO AQUI
m1 = [x[0,40],y[0,40],1]
m2 = [x[0,45],y[0,45],1]
m3 = [x[0,48],y[0,48],1]
m4 = [x[0,54],y[0,54],1]

line_1 = np.cross(m1,m4)
line_2 = np.cross(m2,m3)
print('line_1',line_1)
print('line_2',line_2)

mc = np.cross(line_1,line_2)
xi = mc[0]/mc[2]
yi = mc[1]/mc[2]
print('(xi,yi) = ',(xi,yi))

I = cv2.imread('JohnLennon.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
# Puntos para realizar el cuadrado
quad_x = [x[0,48], x[0,40], x[0,45], x[0,54], x[0,48]]
quad_y = [y[0,48], y[0,40], y[0,45], y[0,54], y[0,48]]

# Graficar la imagen y los landmarks
plt.figure(figsize=(8, 10))
plt.imshow(I)
plt.scatter(xi, yi, c='red')
plt.plot(quad_x, quad_y, c='blue', linewidth=2)  # cuadrilátero

x1 = [m1[0], m4[0]]
y1 = [m1[1], m4[1]]

x2 = [m2[0], m3[0]]
y2 = [m2[1], m3[1]]
plt.plot(x1,y1,color='blue')
plt.plot(x2,y2,color='blue')

plt.scatter(quad_x[:-1], quad_y[:-1], c='red')  # marcar los 4 puntos del cuadrilátero
plt.title('John Lennon - Cuadrilátero ojos , comisura de labios y centro del cuadrilátero')
plt.axis('off')
plt.show()

```

*Encuentre la ecuación de la recta que pasa por mc y la punta de la nariz y grafique esa línea desde mc hasta mq, donde mq es el punto de la línea en x=500.*

*Debe mostrar los cuatro puntos del cuadrilátero, la línea del cuadrilátero, el centro del cuadrilátero, y la línea encontrada, sobre la imagen de John Lennon.*


```python
# Landmark centro del cuadrilátero
m1 = [x[0,40], y[0,40], 1]
m2 = [x[0,45], y[0,45], 1]
m3 = [x[0,48], y[0,48], 1]
m4 = [x[0,54], y[0,54], 1]

# Rectas en homogéneas
line_1 = np.cross(m1, m4)
line_2 = np.cross(m2, m3)
print('line_1', line_1)
print('line_2', line_2)

# Intersección = centro mc
mc = np.cross(line_1, line_2)
xi = mc[0] / mc[2]
yi = mc[1] / mc[2]
print('(xi,yi) = ', (xi, yi))

# Landmark Nariz
nariz = [x[0,30], y[0,30]]

# Pendiente de mc a nariz
m = (nariz[1] - yi) / (nariz[0] - xi)

# Intersección con eje Y
n = yi - m * xi
print(f"Ecuación de la recta: y = {m}x + {n}")

# Calcular mq en x = 500
x_mq = 500
y_mq = m * x_mq + n
print('(x,y) = ', (x_mq, y_mq))

I = cv2.imread('JohnLennon.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

# Graficar la imagen y los landmarks
plt.figure(figsize=(8, 10))
plt.imshow(I)

# Cuadrilátero
plt.plot(quad_x, quad_y, c='blue', linewidth=2)
plt.scatter(quad_x[:-1], quad_y[:-1], s=50, c='red')

# Centro mc
plt.scatter(xi, yi, c='red')

# Nariz
plt.scatter(nariz[0], nariz[1], c='cyan')

# RECTA desde mc hasta mq
plt.plot([xi, x_mq], [yi, y_mq], 'm-', linewidth=2)

# Punto mq
plt.scatter(x_mq, y_mq, c='yellow')

plt.title('John Lennon - Cuadrilátero y dirección mirada')
plt.axis('off')
plt.show()

```
